```python
rent=30000
food=12000
transport=2000
utilities=10000

total_monthly_expenses=rent+food+transport+utilities
budget_limit=60000
print("Total Monthly Expense:", total_monthly_expenses, "PKR")

if total_monthly_expenses <= budget_limit:
    print("You are within the budget limit.")
else:
    print("You have exceeded the budget limit.")
```

```python
attandence=[72, 85, 90, 60, 40, 78, 88]
meet_requirement=0
not_meet=0

for percentage in attandence:
  if percentage>=75 :
    meet_requirement+=1

  else:
    not_meet+=1
print("students meet the attendence requirement (>=75):" , meet_requirement)
print("students not meet the attendence requirment:" , not_meet)
```

```python
cgpas=[3.9 , 3.0 , 3.67, 3.8, 3.5]
incomes=[20000,60000, 30000, 40000, 250000]
eligible_count=0
for i in range(len(cgpas)) :
   if cgpas[i]>=3.5 and incomes[i]<=80000:
     eligible_count+=1
print("total eligible student for scholarship:",eligible_count)
```

```python
reading=[20,56,12,50,90,38,24,59]
min_safe=30
max_safe=50
alert_count=0
for value in reading:
    if value<min_safe or value>max_safe:
        alert_count+=1
print("unsafe reading detection:", value)
total_reading=len(reading)
percentage_unsafe=(alert_count/total_reading)*100
print("\n total alert:",alert_count)
print("percentage of unsafe readings:",percentage_unsafe , "% ")
```
