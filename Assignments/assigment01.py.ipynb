```python

readings = [18, 25, 90, 45, 70, 15, 82, 60]

alert_count = 0
for value in readings:
    if value < 20 or value > 80:
        print(f"Alert! Reading {value} is outside the safe range.")
        alert_count += 1

total_readings = len(readings)

if total_readings > 0:
    alert_percentage = (alert_count / total_readings) * 100
else:
    alert_percentage = 0

print("\nTotal readings:", total_readings)
print("Total alerts:", alert_count)
print(f"Percentage of alert readings: {alert_percentage:.2f}%")
```
